```python
# Funci贸n para limpieza de strings

def normalizar_texto(df):
      for col in df.select_dtypes(include="object").columns:
          df[col] = (
            df[col]
            .astype(str)
            .str.strip()
            .str.replace(r"[\u200b\t\r\n]", "", regex=True)
            .str.replace(" +", " ", regex=True)
            .str.replace("帽", "n", regex=False)
            .str.replace("", "N", regex=False)
            .str.title()
          )

      return df


# Aplicar la funci贸n de normalizaci贸n de string

df_clientes_limpio = normalizar_texto(df_clientes_limpio)
df_ventas_limpio = normalizar_texto(df_ventas_limpio)
df_marketing_limpio = normalizar_texto(df_marketing_limpio)


# Verificar que se hayan normalizado correctamente los dataframes

print("CLIENTES\n\n", df_clientes_limpio.head(10))
print("\n\nVENTAS\n\n", df_ventas_limpio.head(10))
print("\n\nMARKETING\n\n", df_marketing_limpio.head(10))
```

```python
# Normalizar fechas

for df in [df_clientes_limpio, df_ventas_limpio, df_marketing_limpio]:
    for col in df.columns:
        if "fecha" in col.lower():
            df[col] = pd.to_datetime(df[col], errors="coerce", dayfirst=True)


# Verificar que se hayan modificado los tipos de fechas

print("CLIENTES\n\n", df_clientes_limpio.dtypes)
print("\n\nVENTAS\n\n", df_ventas_limpio.dtypes)
print("\n\nMARKETING\n\n", df_marketing_limpio.dtypes)
```

```python
def reporte_calidad_global(dfs, nombres):
    resumen = []
    for df, nombre in zip(dfs, nombres):
        nulos = df.isna().sum().sum()
        duplicados = df.duplicated(keep=False).sum()
        columnas = len(df.columns)
        filas = len(df)
        resumen.append({
            "Dataset": nombre,
            "Filas": filas,
            "Columnas": columnas,
            "Nulos totales": nulos,
            "Duplicados": duplicados,
        })

    reporte = pd.DataFrame(resumen)
    return reporte
```

2. **Transformaci贸n de Datos:** Aplicar filtros y transformaciones para crear una tabla de ventas que muestre solo los productos con alto rendimiento.

=========================

##### 锔 Transformaci贸n de Datos 锔
=========================

```python
# Productos Top 20%

#Funci贸n para detectar columnas

def encontrar_columna(df, candidatos):
    for c in df.columns:
        nombre = c.lower()
        if any(p in nombre for p in candidatos):
            return c
    return None

prod_col = encontrar_columna(df_ventas_limpio, ["producto", "id_producto", "sku", "articulo", "art铆culo"])
if prod_col is None:
    raise ValueError("No se encontr贸 columna de producto. Renombr谩 una columna a 'producto' o similar.")
print(prod_col)
```

```python
# Crear columna ingreso

ventas_perf = (
    df_ventas_limpio
    .assign(
        ingreso = df_ventas_limpio["precio"] * df_ventas_limpio["cantidad"]
    )
)
```

```python
# Res煤men por producto

resumen_prod = (
    ventas_perf
    .groupby(
        by=prod_col,
        dropna=False,
        as_index=False,
        observed=False
    )

    .agg(
        ingreso_total=('ingreso', 'sum'),
        unidades=('cantidad', 'sum'),
        precio_promedio=('precio', 'mean'),
        registros=('ingreso', 'size')
    )
)

# Redondear la columna 'precio_promedio' a 2 decimales
resumen_prod['precio_promedio'] = resumen_prod['precio_promedio'].round(2)

print(resumen_prod.head(10))
```

```python
# Calcular percentil 80 de ingreso_total

p80_ingreso = resumen_prod["ingreso_total"].quantile(
    q=0.80,
    interpolation="linear"
)
```

```python
# Filtrar los productos "de alto rendimiento" y ordenarlos

ventas_top = (
    resumen_prod
    .query(
        "ingreso_total >= @p80_ingreso",
        engine="python"
    )
    .sort_values(
        by=["ingreso_total", "unidades"],
        ascending=[False, False],
        na_position="last",
        ignore_index=True
    )
)

print(f"Columna de producto detectada: {prod_col}")
print(f"Umbral (percentil 80) de ingreso_total: {float(p80_ingreso):,.2f}")
print("\n\nProductos de ALTO RENDIMIENTO (top 20% por ingreso):\n")
print(ventas_top.head(20))
```

3. **Agregaci贸n:** Resumir las ventas por categor铆a de producto y analizar los
ingresos generados.

======================

#####  Agregaci贸n de Datos 
======================

```python
# Resumen por Categor铆a

# Detectar la columna categor铆a
def encontrar_columna(df, candidatos):
    for c in df.columns:
        if any(p in c.lower() for p in candidatos):
            return c
    return None

cat_col = encontrar_columna(df_ventas_limpio, ["categoria", "categor铆a", "categoria_producto", "rubro"])
if cat_col is None:
    raise ValueError("No se encontr贸 columna de categor铆a (por ej. 'categoria' o 'rubro').")

# 2) Asegurar columna 'ingreso' (si no existe, crearla)
if "ingreso" not in df_ventas_limpio.columns:
    ventas_cat = df_ventas_limpio.assign(ingreso = df_ventas_limpio["precio"] * df_ventas_limpio["cantidad"])
else:
    ventas_cat = df_ventas_limpio.copy()

# 3) Agregaci贸n por categor铆a con groupby + agg
resumen_cat = (
    ventas_cat
    .groupby(
        by=cat_col,
        dropna=False,
        as_index=False
    )
    .agg(
        ingreso_total=('ingreso', 'sum'),
        unidades=('cantidad', 'sum'),
        ventas=('ingreso', 'size'),
        precio_promedio=('precio', 'mean')
    )
    .sort_values(
        by='ingreso_total',
        ascending=False,
        na_position='last',
        ignore_index=True
    )
)

# 4) Ticket promedio por venta = ingreso_total / ventas
resumen_cat = resumen_cat.assign(
    ticket_promedio_por_venta = resumen_cat['ingreso_total'] / resumen_cat['ventas']
)

# Redondear la columnas de precios a 2 decimales
columnas_a_redondear = ['precio_promedio', 'ticket_promedio_por_venta']
resumen_cat[columnas_a_redondear] = resumen_cat[columnas_a_redondear].round(2)

print("Columna de categor铆a detectada:", cat_col)
print("\n\nResumen por categor铆a (ordenado por ingreso_total):\n")
print(resumen_cat.head(20))
```

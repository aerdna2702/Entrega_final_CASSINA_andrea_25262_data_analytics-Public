```python
# Copiar los dataframe para no modificar los originales

df_ventas_limpio = df_ventas.copy()
df_clientes_limpio = df_clientes.copy()
df_marketing_limpio = df_marketing.copy()
```

```python
# Eliminar filas duplicadas

df_ventas_limpio = df_ventas_limpio.drop_duplicates()
df_clientes_limpio = df_clientes_limpio.drop_duplicates()
df_marketing_limpio = df_marketing_limpio.drop_duplicates()


# Verificar que se hayan eliminado las filas duplicadas

print("df_ventas_limpio ------> ", df_ventas_limpio.shape)
print("df_clientes_limpio ----> ", df_clientes_limpio.shape)
print("df_marketing_limpio ---> ", df_marketing_limpio.shape)
```

```python
# Funci√≥n para limpieza de strings

def normalizar_texto(df):
      for col in df.select_dtypes(include="object").columns:
          df[col] = (
            df[col]
            .astype(str)
            .str.strip()
            .str.replace(r"[\u200b\t\r\n]", "", regex=True)
            .str.replace(" +", " ", regex=True)
            .str.replace("√±", "n", regex=False)
            .str.replace("√ë", "N", regex=False)
            .str.title()
          )

      return df


# Aplicar la funci√≥n de normalizaci√≥n de string

df_clientes_limpio = normalizar_texto(df_clientes_limpio)
df_ventas_limpio = normalizar_texto(df_ventas_limpio)
df_marketing_limpio = normalizar_texto(df_marketing_limpio)


# Verificar que se hayan normalizado correctamente los dataframes

print("CLIENTES\n\n", df_clientes_limpio.head(10))
print("\n\nVENTAS\n\n", df_ventas_limpio.head(10))
print("\n\nMARKETING\n\n", df_marketing_limpio.head(10))
```

```python
# Normalizar fechas

for df in [df_clientes_limpio, df_ventas_limpio, df_marketing_limpio]:
    for col in df.columns:
        if "fecha" in col.lower():
            df[col] = pd.to_datetime(df[col], errors="coerce", dayfirst=True)


# Verificar que se hayan modificado los tipos de fechas

print("CLIENTES\n\n", df_clientes_limpio.dtypes)
print("\n\nVENTAS\n\n", df_ventas_limpio.dtypes)
print("\n\nMARKETING\n\n", df_marketing_limpio.dtypes)
```

```python
# Normalizar valores num√©ricos

# dataframe df_ventas_limpio:
# columna "precio"

if "precio" in df_ventas_limpio.columns:
   df_ventas_limpio["precio"] = (
        df_ventas_limpio["precio"]
        .astype(str)
        .str.replace("$", "", regex=False)
        .str.replace(",", "", regex=False)
        .str.strip()
    )

   df_ventas_limpio["precio"] = pd.to_numeric(df_ventas_limpio["precio"], errors="coerce")


# Verificar que se hayan eliminado los s√≠mbolos "$"

print('dataframe df_ventas_limpio:\n\ncolumna "precio"\n')
print("\nVENTAS\n\n", df_ventas_limpio.head())
```

##### üìä Reporte Global de Calidad de Datos üìä

```python
def reporte_calidad_global(dfs, nombres):
    resumen = []
    for df, nombre in zip(dfs, nombres):
        nulos = df.isna().sum().sum()
        duplicados = df.duplicated(keep=False).sum()
        columnas = len(df.columns)
        filas = len(df)
        resumen.append({
            "Dataset": nombre,
            "Filas": filas,
            "Columnas": columnas,
            "Nulos totales": nulos,
            "Duplicados": duplicados,
        })

    reporte = pd.DataFrame(resumen)
    return reporte
```

```python
print(reporte_calidad_global(
    [df_clientes, df_ventas, df_marketing],
    ["CLIENTES Original", "VENTAS Original", "MARKETING Original"]
    ))

print("\n",reporte_calidad_global(
    [df_clientes_limpio, df_ventas_limpio, df_marketing_limpio],
    ["CLIENTES Limpio   ", "VENTAS Limpio   ", "MARKETING Limpio   "]
    ))
```

2. **Transformaci√≥n de Datos:** Aplicar filtros y transformaciones para crear una tabla de ventas que muestre solo los productos con alto rendimiento.

=========================

##### üõ†Ô∏è Transformaci√≥n de Datos üõ†Ô∏è
=========================

```python
# Productos Top 20%

# Funci√≥n para detectar columnas

def encontrar_columna(df, candidatos):
    for c in df.columns:
        nombre = c.lower()
        if any(p in nombre for p in candidatos):
            return c
    return None

prod_col = encontrar_columna(df_ventas_limpio, ["producto", "id_producto", "sku", "articulo", "art√≠culo"])
if prod_col is None:
    raise ValueError("No se encontr√≥ columna de producto. Renombr√° una columna a 'producto' o similar.")
print(prod_col)
```

```python
# Crear columna ingreso

df_ingresos_calculados = (
    df_ventas_limpio
    .assign(
        ingreso = df_ventas_limpio["precio"] * df_ventas_limpio["cantidad"]
    )
)
```

```python
# Res√∫men por producto

resumen_prod = (
    df_ingresos_calculados
    .groupby(
        by = prod_col,
        dropna = False,
        as_index = False,
        observed = False
    )

    .agg(
        ingreso_total = ('ingreso', 'sum'),
        unidades = ('cantidad', 'sum'),
        precio_promedio = ('precio', 'mean'),
        registros = ('ingreso', 'size')
    )
)


# Redondear la columna 'precio_promedio' a 2 decimales

resumen_prod['precio_promedio'] = resumen_prod['precio_promedio'].round(2)

print(resumen_prod.head(10))
```

```python
# Calcular percentil 80 de ingreso_total

p80_ingreso = resumen_prod["ingreso_total"].quantile(
    q=0.80,
    interpolation="linear"
)
```

```python
# Filtrar los productos "de alto rendimiento" y ordenarlos

ventas_top = (
    resumen_prod
    .query(
        "ingreso_total >= @p80_ingreso",
        engine = "python"
    )
    .sort_values(
        by = ["ingreso_total", "unidades"],
        ascending = [False, False],
        na_position = "last",
        ignore_index = True
    )
)

print(f"Columna de producto detectada: {prod_col}")
print(f"Umbral (percentil 80) de ingreso_total: {float(p80_ingreso):,.2f}")
print("\n\nProductos de ALTO RENDIMIENTO (top 20% por ingreso):\n")
print(ventas_top.head(20))
```

3. **Agregaci√≥n:** Resumir las ventas por categor√≠a de producto y analizar los
ingresos generados.

======================

##### ‚ûï Agregaci√≥n de Datos ‚ûï
======================

```python
# Resumen por Categor√≠a

# Detectar la columna categor√≠a

def encontrar_columna(df, candidatos):
    for c in df.columns:
        if any(p in c.lower() for p in candidatos):
            return c
    return None

cat_col = encontrar_columna(df_ventas_limpio,
          ["categoria", "categor√≠a", "categoria_producto", "rubro"])
if cat_col is None:
    raise ValueError(
        "No se encontr√≥ columna de categor√≠a (por ej. 'categoria' o 'rubro').")


# Asegurar columna 'ingreso' (si no existe, crearla)

if "ingreso" not in df_ventas_limpio.columns:
    ventas_cat = df_ventas_limpio.assign(
        ingreso = df_ventas_limpio["precio"] * df_ventas_limpio["cantidad"])
else:
    ventas_cat = df_ventas_limpio.copy()


# Agregaci√≥n por categor√≠a con groupby + agg

resumen_cat = (
    ventas_cat
    .groupby(
        by = cat_col,
        dropna = False,
        as_index= False
    )
    .agg(
        ingreso_total = ('ingreso', 'sum'),
        unidades = ('cantidad', 'sum'),
        ventas = ('ingreso', 'size'),
        precio_promedio = ('precio', 'mean')
    )
    .sort_values(
        by='ingreso_total',
        ascending = False,
        na_position = 'last',
        ignore_index = True
    )
)


# Ticket promedio por venta = ingreso_total / ventas

resumen_cat = resumen_cat.assign(
  ticket_promedio_por_venta = resumen_cat['ingreso_total'] / resumen_cat['ventas']
)


# Redondear la columnas de precios a 2 decimales

columnas_a_redondear = ['precio_promedio', 'ticket_promedio_por_venta']
resumen_cat[columnas_a_redondear] = resumen_cat[columnas_a_redondear].round(2)

print("Columna de categor√≠a detectada:", cat_col)
print("\n\nResumen por categor√≠a (ordenado por ingreso_total):\n")
print(resumen_cat.head(20).round(2))
```

4. **Integraci√≥n de Datos:** Combinar los sets de datos de ventas y marketing
para obtener una visi√≥n m√°s amplia de las tendencias.

======================

##### üîó Integraci√≥n de Datos üîó
======================

```python
# Verificar columnas DataFrames para encontrar clave de uni√≥n

print('Columnas de df_ventas_limpio:', df_ventas_limpio.columns)
print('Columnas de df_marketing_limpio:', df_marketing_limpio.columns)
```

Columnas en com√∫n entre ambos DataFrames: 'producto'

```python
# Agrupar df_vetnas_limpio por producto y sumar ingreso

df_ventas_por_producto = df_ventas_limpio.groupby('producto').agg(
    ingreso = ('precio', lambda x: (
        x * df_ventas_limpio.loc[x.index, 'cantidad']).sum())
)


# Ordenar por producto de forma ascendente

df_ventas_por_producto = df_ventas_por_producto.sort_values(
    by = 'producto', ascending = True)

print(df_ventas_por_producto)
```

```python
# Agrupar marketing limpio por producto y sumar costo de campa√±as

df_marketing_por_producto = df_marketing_limpio.groupby('producto')['costo'].sum().reset_index()
df_marketing_por_producto = df_marketing_por_producto[['producto', 'costo']]
df_marketing_por_producto = df_marketing_por_producto.sort_values(by ='producto',
                                                                  ascending=True)

print(df_marketing_por_producto)
```

```python
# Realizar la uni√≥n de los DataFrames

df_ventas_marketing_merge = pd.merge(
    df_ventas_por_producto,
    df_marketing_por_producto,
    on = 'producto'
)

print(df_ventas_marketing_merge)
```

```python
# Calcular % rentabilidad por producto

df_rentabilidad_por_producto = df_ventas_marketing_merge.copy()
df_rentabilidad_por_producto = df_rentabilidad_por_producto.sort_values(
    by = 'producto', ascending = True)
df_rentabilidad_por_producto['%_rentabilidad'] = (
    df_rentabilidad_por_producto['costo'] /
    df_rentabilidad_por_producto['ingreso'] * 100
  ).round(4)

print(df_rentabilidad_por_producto)
```

```python
# Seleccionar columnas "precio", "cantidad"

df_correlacion = df_ventas_limpio[['precio', 'cantidad']].copy()


# Eliminar filas con valores nulos para el c√°lculo

df_correlacion = df_correlacion.dropna()


# Convertir 'cantidad' a tipo num√©rico est√°ndar (float) si es necesario para el c√°lculo

df_correlacion['cantidad'] = df_correlacion['cantidad'].astype(float)

# Mostrar cantidad de filas para el an√°lisis de correlaci√≥n.

print("Filas para el an√°lisis de correlaci√≥n:", len(df_correlacion))
```

```python
# Crear gr√°fico de dispersi√≥n con l√≠nea de regresi√≥n

plt.figure(figsize = (8, 4))
sns.regplot(
    x = 'precio',
    y = 'cantidad',
    data = df_correlacion,
    scatter_kws = {'alpha': 0.6},
    line_kws = {'color': 'red'}
)

plt.title('Relaci√≥n entre Precio y Cantidad Vendida')
plt.xlabel('Precio del Producto')
plt.ylabel('Cantidad Vendida')
plt.grid(True, linestyle = '--', alpha = 0.7)
plt.show()
```

```python
# Calcular la matriz de correlaci√≥n

matriz_correlacion = df_correlacion.corr()


# Extraer el valor espec√≠fico entre 'precio' y 'cantidad'

correlacion_precio_cantidad = matriz_correlacion.loc['precio', 'cantidad']


# Imprimir el resultado

print(f"Coeficiente de Correlaci√≥n (Precio vs. Cantidad): {correlacion_precio_cantidad:.4f}")


if correlacion_precio_cantidad > 0.8:
    print("Interpretaci√≥n: relaci√≥n positiva clara (a mayor precio_promedio, suelen venderse m√°s unidades).")
elif correlacion_precio_cantidad > 0.2:
    print("Interpretaci√≥n: relaci√≥n positiva d√©bil a moderada.")
elif correlacion_precio_cantidad < -0.8:
    print("Interpretaci√≥n: relaci√≥n negativa clara (a mayor precio_promedio, suelen venderse menos unidades).")
elif correlacion_precio_cantidad < -0.3:
    print("Interpretaci√≥n: relaci√≥n negativa d√©bil a moderada.")
else:
    print("Interpretaci√≥n: relaci√≥n muy d√©bil o nula.")

```
